```python
import MeCab

# 下記のパスは、環境設定時に表示された値にすること
tagger = MeCab.Tagger('-d /usr/lib/x86_64-linux-gnu/mecab/dic/mecab-ipadic-neologd')

# withを使用してテキストファイルを開く
with open('gongitsune.txt', encoding = 'shift-jis') as f:
    # 1行ずつ読み込む
    line = f.readline()

    print(line)

    # テキストファイルの最終行までループする
    # lineにデータが存在する間ループする
    while line:
        text = line.strip()

        tagger.parse('')
        # 形態素解析を行う
        node =  tagger.parseToNode(text)

        while node is not None:
            # 形態素解析の結果から、品詞だけ取り出す
            hinshi = node.feature.split(",")[0]

            # 形態素解析の結果から、形態素だけ取り出す
            word = node.surface

            if  hinshi not in ['BOS/EOS']:
                # 出力例どおりに出力する
                print('形態素：{} 品詞：{}'.format(word, hinshi))

            # 形態素解析の結果の次のノードを読み込む
            node = node.next
     
        # 次の行を読み込む
        line = f.readline()

        print(line)
```
